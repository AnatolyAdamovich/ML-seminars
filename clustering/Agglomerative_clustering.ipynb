# libraries

```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt
from custom_agglomerative import AgglomerativeAlgorithm

%matplotlib inline
```

# data

```python
data = np.array([[0, 0], [2, 4], [3, 3], [1, 2], [3, 0], [3, 1], [1, 1], [12, 18], [13, 17], [11, 15], [13, 14],
                 [14, 16], [11, 16], [12, 15], [13, 18], [12, 5], [13, 2], [14, 4], [12, 3], [13, 1], [14, 2], [24, 19],
                 [22, 22],  [21, 24], [23, 21], [24, 20], [22, 39], [23, 38], [24, 39], [21, 37], [2, 26], [24, 6], [10, 36]])

data.shape
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(data[:, 0], data[:, 1])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Исходный набор данных')
```

# baseline

```python
sklearn_model = AgglomerativeClustering(n_clusters=4, linkage='single').fit(data)
scipy_model = hierarchy.linkage(y=data, method='single')
```

```python
sklearn_model.labels_
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(data[:, 0], data[:, 1], c=sklearn_model.labels_)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Исходный набор данных')
```

```python
print(scipy_model)
```

```python
fig = plt.figure(figsize=(25, 10))
dn = hierarchy.dendrogram(scipy_model)
plt.show()
```
